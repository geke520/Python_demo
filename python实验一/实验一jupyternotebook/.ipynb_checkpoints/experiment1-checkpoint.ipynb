```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
df  = pd.DataFrame(recent_grads)
print("数据的第一行为:\n",df.iloc[0])
#输出CSV文件的前5行和最后5行，pandas默认的输出5行
print("\n数据的前5行为:\n",df.head())
print("\n数据的前5行为:\n",df.tail())
print("\n数据摘要统计:\n",df.describe())
```

```python
raw_data_count = df.iloc[:,0].size
print(raw_data_count)#打印行数
```

```python
recent_grads = pd.DataFrame.dropna(recent_grads)#删除缺失值
```

```python
cleaned_data_count = recent_grads.iloc[:,0].size
print(cleaned_data_count)#打印行数
```
